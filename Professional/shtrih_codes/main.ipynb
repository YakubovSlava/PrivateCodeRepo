```python
im = Image.open('ABBYY_dataset/Test/Segmaps/0001.tif')
```

```python
im = abs((im-255))
```

```python
im = cv2.imread('ABBYY_dataset/Test/Segmaps/0002.tif')
plt.imshow(im)
```

```python
# closed = cv2.erode(im, None, iterations = 10)
closed = cv2.dilate(im, None, iterations = 30)
```

```python
plt.imshow(closed)
```

```python
imgray = cv2.cvtColor(closed, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
plt.figure(figsize=(30, 30))
plt.imshow(cv2.drawContours(im, contours, -1, (0,255,0), 3))
```
