```python
import torch
```

```python
from PIL import Image
```

```python
img = Image.open('D:\\mlp\\MachineLearning\\detection2\\017.jpg')
```

```python
img
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Inference
results = model(img, size=640, force_reload=True)

# Results
results.print()
results.save()  # or .show()
```
